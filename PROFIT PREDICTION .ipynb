```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from xgboost import XGBRFRegressor
import seaborn as sns
from sklearn import metrics
```

```python
df = pd.read_csv(r"C:\Users\PC\Desktop\MACHINE LEARNING\1000_Companies.csv")
df
```

```python
df.isnull().sum()
```

```python
df['State'].value_counts ()
```

```python
df.replace({'State':{'California':0,'New York':1,'Florida':2}},inplace=True)
```

Separating Data and Labels

```python
X = df.drop(columns='Profit')
Y = df['Profit']
```

Training-data 

```python
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size=0.2,random_state=2)
```

Model Training 

```python
model = LinearRegression ()
```

```python
model.fit(X_train,Y_train)
```

Model Evaluation

```python
#Accuracy on training data
Training_data_prediction = model.predict (X_train)

# R squared error
score_1 = metrics.r2_score(Y_train,Training_data_prediction)

# mean absolute value
score_2= metrics.mean_absolute_error(Y_train,Training_data_prediction)

print('R Squared R :', score_1)
print('Mean Absolute Error :', score_2)
```

Visualization

```python
plt.scatter(Y_train,Training_data_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Acutal Price Vs Predicted Price')
plt.show()
```

```python
#Accuracy on test data
Test_data_prediction = model.predict (X_test)

# R squared error
score_1 = metrics.r2_score(Y_test,Test_data_prediction)

# mean absolute value
score_2= metrics.mean_absolute_error(Y_test,Test_data_prediction)

print('R Squared R :', score_1)
print('Mean Absolute Error :', score_2)
```

Visualization

```python
plt.scatter(Y_test,Test_data_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Acutal Price Vs Predicted Price')
plt.show()
```
