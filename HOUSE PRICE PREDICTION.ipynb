# BOSTON HOUSE PRICE PREDICTION USING XGBoost Regressor

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
df = pd.read_csv(r"C:\Users\PC\Desktop\MACHINE LEARNING\boston.csv")

df.rename(columns = {'MEDV':'PRICE'},inplace=True)

df
```

Finding Correlation

```python
correlation = df.corr()
#Plotting Heat Map of correlation 
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(), cbar=True, square= True, fmt='.1f', annot=True, annot_kws={'size':10}, cmap='Greens')
```

Separating data and Labels

```python
X=df.drop(['PRICE'],axis=1)
Y=df['PRICE']
```

```python
print(X)
print(Y)
```

Spiliting Data into Training and Test data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
X_train.shape,X_test.shape
```

Model Training

```python
#Loading the model
model=XGBRegressor()
```

```python
#Training the model
model.fit(X_train,Y_train)
```

Evaluation 

Training Data Accuracy 

```python
#Accuracy on training data
Training_data_prediction = model.predict (X_train)
print(Training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train,Training_data_prediction)

# mean absolute value
score_2= metrics.mean_absolute_error(Y_train,Training_data_prediction)

print('R Squared R :', score_1)
print('Mean Absolute Error :', score_2)
```

Visualizing Actual Data and Predicted Data

```python
plt.scatter(Y_train,Training_data_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Acutal Price Vs Predicted Price')
plt.show()
```

Test Data Acccuracy 

```python
#Accuracy on test data
Test_data_prediction = model.predict (X_test)
print(Test_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_test,Test_data_prediction)

# mean absolute value
score_2= metrics.mean_absolute_error(Y_test,Test_data_prediction)

print('R Squared R :', score_1)
print('Mean Absolute Error :', score_2)
```

Visualization of Actual and Predicted data

```python
plt.scatter(Y_test,Test_data_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Acutal Price Vs Predicted Price')
plt.show()
```
